```python
import pandas as pd 


cashflow_df = pd.read_csv('data/cashflow.csv')
cashflow_df['Escola'] = cashflow_df['Escola'].replace('80', 'IEI SÃO VICENTE DE PAULO')
cashflow_df['Escola'] = cashflow_df['Escola'].str.strip()
aggregated_cashflow_df = cashflow_df.groupby('Escola')['Valor'].sum().reset_index()
aggregated_cashflow_df.to_csv('data/aggregated_cashflow.csv', index=False)
```
